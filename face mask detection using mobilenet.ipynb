```python
dataset_path = 'D:\\data'
categories = ['with_mask', 'without_mask']
data = []

for category in categories:
    category_path = os.path.join(dataset_path, category)
    label = categories.index(category)
    
    for file in os.listdir(category_path):
        image_path = os.path.join(category_path, file)
        img = cv2.imread(image_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (224, 224))
        data.append([img, label])
```

```python
import random

random.shuffle(data)
```

```python
X=[]
y=[]
for features,label in data:
    X.append(features)
    y.append(label)
```

```python
                        
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Input
from tensorflow.keras.models import Model


base_model = MobileNetV2(weights='imagenet', include_top=False, input_tensor=Input(shape=(224, 224, 3)))

base_model_output = base_model.layers[-2].output

x = GlobalAveragePooling2D()(base_model_output)


for layer in base_model.layers:
    layer.trainable = False



predictions = Dense(1, activation='sigmoid')(x)

new_model = Model(inputs=base_model.input, outputs=predictions)

for layer in new_model.layers[-2:]:
    layer.trainable = True

new_model.summary()
```

```python
from tensorflow.keras.optimizers import Adam

new_model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
    new_model.fit(X_train, y_train,epochs=5,validation_data=(X_test, Y_test))
```

```python
import cv2
import numpy as np
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input


model = new_model  


face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)

if model is None:
    raise ValueError("Model is not defined. Please load the model before running the script.")

while True:

    ret, frame = cap.read()

    if not ret:
        break


    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    
    faces = face_cascade.detectMultiScale(rgb_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
      
        face_roi = rgb_frame[y:y+h, x:x+w]
        face_roi = cv2.resize(face_roi, (224, 224))
        face_roi = np.expand_dims(face_roi, axis=0)

        
        face_roi = preprocess_input(face_roi)

        
        prediction = model.predict(face_roi)


        if prediction[0][0] > 0.5:
            label = "Without Mask"
            color = (0, 0, 255) 
        else:
            label = "With Mask"
            color = (0, 255, 0) 

       
        cv2.putText(frame, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
        cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)


    cv2.imshow('Window', frame)


    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()

```
